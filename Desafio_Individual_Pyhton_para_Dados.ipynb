8. Quem tem maior pressão sanguínea média, homens ou mulheres?
(Considere a média)

```python
# Função para converter o formato de pressão sanguínea para números
def converter_pressao(pressao):
    try:
        sistolica, diastolica = map(int, pressao.split('/'))
        return sistolica, diastolica
    except ValueError:
        return None, None
```

```python
# Aplicando a função para criar novas colunas de pressão sistólica e diastólica
df_saude[['Pressão Sistólica', 'Pressão Diastólica']] = df_saude['Pressão sanguínea'].apply(lambda x: pd.Series(converter_pressao(x)))
```

```python
# Calculando a média de pressão sanguínea total (sistólica + diastólica) para homens
pressao_total_homens = df_saude[df_saude['Gênero'] == 'Homem'][['Pressão Sistólica', 'Pressão Diastólica']].sum(axis=1).mean()

# Calculando a média de pressão sanguínea total (sistólica + diastólica) para mulheres
pressao_total_mulheres = df_saude[df_saude['Gênero'] == 'Mulher'][['Pressão Sistólica', 'Pressão Diastólica']].sum(axis=1).mean()
```

```python
# Comparando as médias e exibindo resultado com print
if pd.notna(pressao_total_homens) and pd.notna(pressao_total_mulheres):
    if pressao_total_homens > pressao_total_mulheres:
        resultado = "Homens têm a maior pressão sanguínea média."
    elif pressao_total_mulheres > pressao_total_homens:
        resultado = "Mulheres têm a maior pressão sanguínea média."
    else:
        resultado = "Homens e mulheres têm a mesma pressão sanguínea média."
else:
    resultado = "Não foi possível calcular as médias devido a falta de dados válidos."
print(resultado)
```

9. É predominante entre os participantes dormir 8 horas por dia
(considere usar Moda como medida)?

```python
# Calculando a moda da coluna 'Duração do sono' e exibindo resultado
moda_duracao_sono = df_saude['Duração do sono'].mode()
moda_duracao_sono
```

10. Pessoas com frequências cardíacas acima de 70 dão mais passos que
pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
# Cálculando a média de passos diários para frequência cardíaca acima de 70 e exibindo o resultado
media_passos_acima_70 = df_saude[df_saude['Frequência cardíaca'] > 70]['Passos diários'].mean()
media_passos_acima_70
```

```python
# Cálculando a média de passos diários para frequência cardíaca menor ou igual 70 e exibindo o resultado
media_passos_70_ou_menos = df_saude[df_saude['Frequência cardíaca'] <= 70]['Passos diários'].mean()
media_passos_70_ou_menos
```

```python
# Comparando as médias e exibindo resultado com o print
if media_passos_acima_70 > media_passos_70_ou_menos:
    print("Sim, pessoas com frequência cardíaca acima de 70 dão mais passos.")
elif media_passos_acima_70 < media_passos_70_ou_menos:
    print("Não, pessoas com frequência cardíaca igual ou menor a 70 dão mais passos.")
else:
    print("Não, pessoas com frequência cardíaca acima de 70 e Pessoas com frequência cardíaca igual ou menor a 70 dão o mesmo número de passos.")
```
